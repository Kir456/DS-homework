```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#загрузка датафреймов
dau = pd.read_excel('hw_lesson03_dau.xlsx')
dau_device = pd.read_excel('hw_lesson03_dau_device.xlsx')
dau_channel = pd.read_excel('hw_lesson03_dau_channel.xlsx')
dau_region = pd.read_excel('hw_lesson03_dau_region.xlsx')
```

```python
#форматирование данных(отображение даты, перенос индекса)
dau['date'] = pd.to_datetime(dau['date'])
dau = dau.set_index('date')

dau_channel['date'] = pd.to_datetime(dau_channel['date'])
dau_channel = dau_channel.set_index('date')

dau_device['date'] = pd.to_datetime(dau_device['date'])
dau_device = dau_device.set_index('date')

dau_region['date'] = pd.to_datetime(dau_region['date'])
dau_region = dau_region.set_index('date')
```

```python
plt.rcParams.update({'font.size': 8})
plt.plot(dau.index, dau)
plt.xlabel('Дата')
plt.ylabel('Количество пользователей')
plt.title('Общие данные')
plt.show()
```

```python
# По графику заметно, что с минимального количества пользователей в первые дни показатели к концу стремительно выросли
# Также можно отметить, что этот рост был неравномерный
```

```python
plt.plot(dau_device.index, dau_device)
plt.legend(dau_device.columns.tolist())
plt.xlabel('Дата')
plt.ylabel('Количество пользователей')
plt.title('Устройства')
plt.show()
```

```python
# По графику можно заметить, что в несколько раз количество людей, пользующихся android устройствами, больше, чем пользователей iphone
```

```python
plt.plot(dau_region.index, dau_region)
plt.legend(dau_region.columns.tolist())
plt.xlabel('Дата')
plt.ylabel('Количество пользователей')
plt.title('Регионы')
plt.show()
```

```python
# По графику можно заметить, что для каждого региона сохранялась тенденция неравномерного изменения количества пользователей
# Также каждый регион за всё время почти что не менял свою "позицию" в общем рейтинге 
```

```python
plt.plot(dau_channel.index, dau_channel)
plt.legend(dau_channel.columns.tolist())
plt.xlabel('Дата')
plt.ylabel('Количество пользователей')
plt.title('Каналы')
plt.show()
```

```python
# По графику можно заметить, что за всё время сохранялось примерно одинаковое количество пользоваталей у каждого канала
# Правда стоит отметить, что сначала у GameAds были чуть выше показатели, чем у JetAds, но к концу тенденция изменилась в обратную сторону
```
