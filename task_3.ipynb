```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#создаем 2 датафрейма по 2 файлам
df_tutorial = pd.read_csv('hw_lesson02_tutorial_completed.csv')
df_sessions = pd.read_excel('hw_lesson02_sessions.xlsx')
```

```python
#диаграмма размаха для 1 датафрейма
plt.boxplot(x=df_tutorial['duration_sec'])
```

```python
#диаграмма размаха для 2 датафрейма
plt.boxplot(x=df_sessions['session_duration'])
```

```python
#99% перцентиль 1 датафрейма
df_tutorial['duration_sec'].quantile(0.99)
```

```python
#99% перцентиль 2 датафрейма
df_sessions['session_duration'].quantile(0.99)
```

```python
#по данной статистике можно заметить, что присутствует достаточно большое количество "выбросов", которые причем входят в 1% данных
```
