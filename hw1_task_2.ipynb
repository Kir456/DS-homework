```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#создаем 2 датафрейма по 2 файлам
df_tutorial = pd.read_csv('hw_lesson02_tutorial_completed.csv')
df_sessions = pd.read_excel('hw_lesson02_sessions.xlsx')
```

```python
#среднее значение длительности по 1 датафрейму
df_tutorial['duration_sec'].mean()
```

```python
#среднее значение длительности по 2 датафрейму
df_sessions['session_duration'].mean()
```

```python
#среднеквадратическое отклонение длительности по 1 датафрейму
df_tutorial['duration_sec'].std()
```

```python
#среднеквадратическое отклонение длительности по 2 датафрейму
df_sessions['session_duration'].std()
```

```python
#медиана длительности по 1 датафрейму
df_tutorial['duration_sec'].median()
```

```python
#медиана длительности по 2 датафрейму
df_sessions['session_duration'].median()
```

```python
#среднее значение длительности по 1 датафрейму в минутах
df_tutorial['duration_sec'].mean()/60
```

```python
#среднее значение длительности по 2 датафрейму в минутах
df_sessions['session_duration'].mean()/60
```

```python
#медиана длительности по 1 датафрейму в минутах
df_tutorial['duration_sec'].median()/60
```

```python
#медиана длительности по 2 датафрейму в минутах
df_sessions['session_duration'].median()/60
```

```python
#вывод гистограммы с продолжительностью по 1 датафрейму
plt.plot(df_tutorial.index, df_tutorial['duration_sec'])
plt.show()
```

```python
#вывод гистограммы с продолжительностью по 2 датафрейму
plt.plot(df_sessions.index, df_sessions['session_duration'])
plt.show()
```

```python
#ну про средние, медианные и среднеквадратичные значения сложно что-то сказать(ну только можно отметить, что медиана сильно, но не прям уж, отличается от среднего значения);
#по гистограммам же можно отметить, что в 1 датафрейме преимущественно все значения примерно одинаковые, кроме некоторых редких "скачков", а во 2 датафрейме в целом всё примерно одинаковое
```
